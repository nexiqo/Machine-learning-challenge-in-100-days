```python
class fraction(object):
    def __init__(self,num,den):
        self.num=num
        self.den=den

    def __lt__(self,other):
        return (self.num / self.den) < (other.num / other.den)

    def __add__(self,other):
        # For addition: a/b + c/d = (ad + bc) / bd
        new_num = (self.num * other.den) + (other.num * self.den)
        new_den = self.den * other.den
        return fraction(new_num, new_den)

    def __mul__(self,other):
        # For multiplication: a/b * c/d = (ac) / (bd)
        new_num = self.num * other.num
        new_den = self.den * other.den
        return fraction(new_num, new_den)

f1=fraction(1,2)
f2=f1.__mul__(f1) # Using __mul__ directly or f1 * f1 if operators are overloaded
# To test the __lt__ method, f2=f1.mul(f1) is 1/4. f1 is 1/2.
# So f1 < f2 should be False (1/2 < 1/4 is False)
# f1 > f2 should be True

# Let's verify with explicit fraction values for clearer test
f1_test = fraction(1,2)
f2_test = fraction(1,4) # Assuming f1.mul(f1) would result in 1/4 if f1 is 1/2

print(f1_test.__lt__(f2_test)) # This should print False (1/2 < 1/4 is False)
```
