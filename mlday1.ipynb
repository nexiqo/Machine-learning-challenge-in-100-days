```python
import matplotlib.pyplot as plt
words=['hello','worlds','Iam','Computer']
freq=[5,6,1,8]
plt.bar(words,freq)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
c=['a','b','c','d','e']
a=np.array([1,2,3,4,5])
b=np.ones(5)
plt.ylim(0,7)
plt.bar(c,a+b)
plt.show()
print(a)
print(a+b)
```

```python
import matplotlib.pyplot as plt
xpoints=[1,2,4,8]
ypoints=[3,6,1,15,8,9]#x goes in acending order
plt.plot(ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
ypoints=[1,3,5,7,8]
plt.plot(ypoints,marker='o',ms='10')#ms = size of marker
plt.show()
```

```python
import matplotlib.pyplot as plt
ypoints=[2,3,5,6,9]
plt.plot(ypoints,marker='o',mec='r',ms=10)#mec=border colour of marker
```

```python
import matplotlib.pyplot as plt
ypoints=[1,2,3,4,5,6]
plt.plot(ypoints,marker='o',mfc='r',ms=10)
```
