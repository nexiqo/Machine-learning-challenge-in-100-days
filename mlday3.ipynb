```python
import matplotlib.pyplot as plt
import numpy as np

times=[6,12,18,24]

def plot_mood_by_time(moods):
  x_values = []
  for x in range(int(len(moods)/4)):
    x_values.append(x)
  for t in times:
    result=[]
    for input_item in moods:
      if input_item[2]==t:
        result.append(input_item[3])
    plt.plot(np.array(x_values),np.array(result), label = 'time'+str(t))
    plt.ylim(0,10)
    plt.xlim(0,len(x_values))
  plt.show()

# Sample data for 'moods' to fix the NameError
# Each inner list represents a data point, with index 2 being the time and index 3 being the mood score
moods = [
    [1, '2023-01-01', 6, 7],    # Day 0, Time 6, Mood 7
    [2, '2023-01-01', 12, 5],   # Day 0, Time 12, Mood 5
    [3, '2023-01-01', 18, 8],   # Day 0, Time 18, Mood 8
    [4, '2023-01-01', 24, 6],   # Day 0, Time 24, Mood 6
    [5, '2023-01-02', 6, 6],    # Day 1, Time 6, Mood 6
    [6, '2023-01-02', 12, 7],   # Day 1, Time 12, Mood 7
    [7, '2023-01-02', 18, 9],   # Day 1, Time 18, Mood 9
    [8, '2023-01-02', 24, 7]    # Day 1, Time 24, Mood 7
]

plot_mood_by_time(moods)
```
